```python
import pandas as pd
import seaborn as sns

data =sns.load_dataset('titanic')
```

```python
data.head()
```

```python
data['pclass'].unique()
```

<h3>Preprocessing The Datas</h3>

```python
cols_to_drop= ['embarked','class','who','adult_male','deck','embark_town','alive','alone']

data.drop(columns=cols_to_drop,inplace=True)
data.head()
```

<h3> Now Normalizing the dataaset</h3>

```python
numerial_cols =['age','sibsp','parch','fare']

data[numerial_cols] = (data[numerial_cols] - data[numerial_cols].min())/ (data[numerial_cols].max() - data[numerial_cols].min())

data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoding = LabelEncoder()

data['sex'] = lbl_encoding.fit_transform(data['sex'])

data.head()
```

```python
rows_before = data.shape[0]

print('before' , rows_before)

data.dropna(inplace= True)

rows_after = data.shape[0]

print('rows after dropping the NAN values ', rows_after)

data.head()
```

```python
from sklearn.model_selection import train_test_split

x= data[['pclass','sex','age','sibsp','parch','fare']]

# x= data.drop('Survived',axis=1)

y = data['survived']

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state= 1)


print('Value of x_train shape',x_train.shape)
print('Value of x_test shape',x_test.shape)
print('Value of y_train shape',y_train.shape)
print('Value of x_test shape',y_test.shape)


#issue Solved 

```

<h3>Applying KNN model to predict the survival of passengers</h3>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

k = int(input('Enter the value of K :'))

knn_model = KNeighborsClassifier (n_neighbors=k)

knn_model.fit(x_train,y_train)

predicitions = knn_model.predict(x_test)

detailed_rpt = classification_report(predicitions,y_test)

accu = accuracy_score(predicitions,y_test)


print('Actual Accuracy ---> ',accu*100,'(%) \n')

print('Detailed Report :\n',detailed_rpt)

```
